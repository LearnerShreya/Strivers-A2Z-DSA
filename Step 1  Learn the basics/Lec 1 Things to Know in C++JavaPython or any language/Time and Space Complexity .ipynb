What is Time Complexity?  
The rate at which the time, required to run a code, changes with respect to the input size, is considered the time complexity. Basically, the time complexity of a particular code depends on the given input size, not on the machine used to run the code.


Here come the three rules, that we are going to follow while calculating the time complexity:
- We will always calculate the time complexity for the worst-case scenario.
- We will avoid including the constant terms.
- We will also avoid the lower values.

What is Space Complexity?  
Space complexity generally represents the summation of auxiliary space and the input space. Auxiliary space refers to the space that we use additionally to solve a problem. And input space refers to the space that we use to store the inputs.

which represents the sum of the first \( n \) natural numbers:

$$
1 + 2 + 3 + 4 + \dots + n

Here’s a way to **visualize it step by step**:

### **1. Pairing Method (Triangular Numbers)**
- Imagine writing the sequence forward:
  $$
  1 + 2 + 3 + \dots + (n-1) + n
  $$
- Now, write the same sequence **backward**:
  $$
  n + (n-1) + (n-2) + \dots + 2 + 1
  $$

- If you **add both sequences**, each column sums to \( (n+1) \):

  $$
  (1+n) + (2 + (n-1)) + (3 + (n-2)) + \dots + ((n-1)+2) + (n+1)
  $$

- There are **n pairs**, each summing to $ (n+1) $.

- So, the total sum of both sequences is:
  $$
  n(n+1)
  $$

- Since we counted everything **twice**, the actual sum is:

  $$
  S = \frac{n(n+1)}{2}
  $$

### **2. Visual Representation (Triangle Formation)**
Think of arranging dots in a **triangle**:

- For $ n = 5 $:

  ```
      ●
     ● ●
    ● ● ●
   ● ● ● ●
  ● ● ● ● ●
  ```

- The **total number of dots** is the sum $ 1+2+3+4+5 $.

- If you **duplicate** this triangle and rotate it, it forms a **rectangle** of size $ n \times (n+1) $, so the total area is:

  $$
  n(n+1)
  $$

- Since we took **double**, we divide by 2:

  $$
  S = \frac{n(n+1)}{2}
  $$

## **📌 Understanding Time Complexity and Execution Limits**  
Most online competitive programming platforms execute **$10^8$ operations per second**.  
If the **time limit is 1 second**, your solution should stay **within $O(10^8)$** operations to avoid **TLE (Time Limit Exceeded)** errors.  

---

## **🛠️ Time Complexity Guide**  
| **Time Complexity** | **Max $n$ in 1s** | **Example Algorithm** |  
|--------------------|-----------------|----------------|  
| $O(1)$ (Constant Time) | Any $n$ | Direct calculations |  
| $O(\log n)$ (Logarithmic Time) | $n \approx 10^{18}$ | **Binary Search** |  
| $O(n)$ (Linear Time) | $n \approx 10^8$ | **Single loop, array traversal** |  
| $O(n \log n)$ (Linearithmic Time) | $n \approx 10^6$ | **Sorting algorithms** |  
| $O(n^2)$ (Quadratic Time) | $n \approx 10^4$ | **Nested loops** |  
| $O(n^3)$ (Cubic Time) | $n \approx 500$ | **Triple nested loops** |  
| $O(2^n)$ (Exponential Time) | $n \approx 20$ | **Backtracking** |  
| $O(n!)$ (Factorial Time) | $n \approx 10$ | **Permutations** |  

---

## **🔍 Breakdown with Python Code**  
### **1️⃣ Constant Time - $O(1)$**  
**✅ Fastest** (No loops, direct computation).  
- Example: **Sum of two numbers**  
  ```python
  def add_numbers(a, b):
      return a + b  # Always executes in 1 step
  ```
- **Runs instantly regardless of $n$.**  

---

### **2️⃣ Logarithmic Time - $O(\log n)$**  
**✅ Efficient for large $n$**  
- Example: **Binary Search**  
  ```python
  def binary_search(arr, target):
      left, right = 0, len(arr) - 1
      while left <= right:
          mid = (left + right) // 2
          if arr[mid] == target:
              return mid
          elif arr[mid] < target:
              left = mid + 1
          else:
              right = mid - 1
      return -1
  ```
- Can handle **$n \approx 10^{18}$** efficiently.  

---

### **3️⃣ Linear Time - $O(n)$**  
**✅ Works well for $n \leq 10^8$**  
- Example: **Finding sum of an array**  
  ```python
  def sum_array(arr):
      total = 0
      for num in arr:
          total += num
      return total
  ```
- If $n = 10^8$, it **just fits** within 1 second.  

---

### **4️⃣ Linearithmic Time - $O(n \log n)$**  
**✅ Sorting Algorithms (Merge Sort, QuickSort)**  
- Example: **Sorting an array**  
  ```python
  arr.sort()  # Python uses Timsort (O(n log n))
  ```
- Can handle **$n \approx 10^6$** efficiently.  
- **If $n = 10^8$, this complexity is too slow.**  

---

### **5️⃣ Quadratic Time - $O(n^2)$**  
**🚨 Avoid for large $n$ 🚨**  
- Example: **Brute-force nested loops**  
  ```python
  def pairwise_combinations(arr):
      n = len(arr)
      for i in range(n):
          for j in range(n):
              print(arr[i], arr[j])
  ```
- Can handle only **$n \approx 10^4$**  
- **If $n = 10^6$, it would take $10^{12}$ operations → TLE!**  

---

### **6️⃣ Cubic Time - $O(n^3)$**  
**🚨 Only works for small $n$ (~500 max) 🚨**  
- Example: **Triple nested loop**  
  ```python
  def matrix_operations(matrix):
      n = len(matrix)
      for i in range(n):
          for j in range(n):
              for k in range(n):
                  matrix[i][j] += matrix[j][k]  
  ```
- If $n = 500$, it's **just under** $10^8$ operations.  
- **If $n = 1000$, it exceeds the limit → TLE!**  

---

### **7️⃣ Exponential Time - $O(2^n)$**  
**🚨 Works only for $n \leq 20$ 🚨**  
- Example: **Recursive Fibonacci (very slow!)**  
  ```python
  def fibonacci(n):
      if n <= 1:
          return n
      return fibonacci(n - 1) + fibonacci(n - 2)
  ```
- **For $n = 40$, it takes $10^{12}$ operations → Too slow!**  

---

### **8️⃣ Factorial Time - $O(n!)$**  
**🚨 Feasible only for small $n$, $n \leq 10$**  
- Example: **Generating all permutations**  
  ```python
  from itertools import permutations
  
  def all_permutations(arr):
      return list(permutations(arr))
  ```
- **For $n = 10$, it takes $10! = 3.6 \times 10^6$ operations.**  
- **For $n = 15$, it takes $10^{12}$ operations → TLE!**  

---

## **🎯 Key Takeaways**  
✅ **Use $O(n \log n)$ or better for large inputs.**  
✅ **$O(n^2)$ is bad for $n > 10^4$, and $O(2^n)$ is impractical for $n > 20$.**  
✅ **For $n \approx 10^6$, prefer sorting ($O(n \log n)$) or linear approaches ($O(n)$).**  
✅ **Factorial and exponential time complexities should be avoided unless $n$ is very small.**
