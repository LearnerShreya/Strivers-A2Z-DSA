```python
class Solution:
    def dataTypeSize(self, str):
        # Code here
        sizes = {
            "Character": 1,
            "Integer": 4,
            "Long": 8,
            "Float": 4,
            "Double": 8
        }
        return sizes.get(str, -1)

sol = Solution()
sol.dataTypeSize("Float")
```

Decision Making in Java  
https://www.geeksforgeeks.org/problems/java-if-else-decision-making0924/0?category%255B%255D=Java&difficulty%255B%255D=-2&page=1&query=category%255B%255DJavadifficulty%255B%255D-2page1category%255B%255DJava  

```python
class Solution:
    def compareNM(self, n: int, m: int) -> str:
        if n < m:
            return "lesser"
        elif n == m:
            return "equal"
        else:
            return "greater"


# class Solution:
#     def compareNM(self, n: int, m: int) -> str:
#         return "lesser" if n < m else "equal" if n == m else "greater"

# Create an instance of the Solution class
obj = Solution()

# Call the compareNM method
result = obj.compareNM(2, 3)

# Print the output
print(result)  # Output: lesser
```

```python
class Solution:
    @staticmethod
    def compareNM(n: int, m: int) -> str:
        if n < m:
            return "lesser"
        elif n == m:
            return "equal"
        else:
            return "greater"

# Now, you can call it directly without creating an instance
print(Solution.compareNM(2, 3))  # Output: lesser
```
