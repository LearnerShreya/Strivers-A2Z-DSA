````python
class Solution:
    def printSquare(self, N):
        for i in range(N):  # Loop for N lines
            for j in range(N):  # Loop for N numbers in each line
                print(N, end=" ")  # Print N with space
            print()  # Move to the next line after printing N times

# Example usage:
sol = Solution()
sol.printSquare(4)
````



Let's break down both implementations step by step.

---

## **🔹 Concept Behind First Code (`printSquare(N)`)**
### **Code:**
```python
class Solution:
    def printSquare(self, N):
        for i in range(N):  # Loop for N lines
            for j in range(N):  # Loop for N numbers in each line
                print(N, end=" ")  # Print N with space
            print()  # Move to the next line after printing N times

# Example usage:
sol = Solution()
sol.printSquare(4)
```

### **🔍 Explanation:**
✅ This function prints a square pattern of size `N × N` using the number `N`.  
✅ **Nested Loops**:  
   - **Outer loop (`for i in range(N)`)** → Controls the number of rows (`N` rows).  
   - **Inner loop (`for j in range(N)`)** → Controls the number of elements printed in each row (`N` times).  
✅ **Printing with `end=" "`**:  
   - By default, `print()` adds a **newline (`\n`)** at the end of each statement.  
   - `end=" "` ensures numbers are printed on the same line, separated by a space.  
✅ **Newline after each row**:  
   - The `print()` statement after the inner loop moves to the next line.

## **🔹 Concept Behind Second Code (`printSquare(n)`)**
### **Code:**
```python
class Solution:
    def printSquare(self, n):
        for i in range(n):  # Loop for n rows
            print("* " * n)  # Print n stars with space
            
# Example usage:
sol = Solution()
sol.printSquare(3)
```

### **🔍 Explanation:**
✅ This function prints a square pattern of size `n × n` using `*`.  
✅ **Single Loop** (`for i in range(n)`)  
   - Since we only need `n` rows, we use one loop instead of two.  
✅ **Using `"* " * n"` for Printing**  
   - `"* " * n` creates a string with `n` stars separated by spaces.
   - This reduces the need for a second loop, making the code **shorter and more efficient**.

### **🔹 Dry Run (`n = 3`)**
```
n = 3

Outer Loop Iteration (i) | Output
-----------------------------------
i = 0                    | * * *
i = 1                    | * * *
i = 2                    | * * *
```

### **🔹 Output (`n = 3`)**
```
* * *
* * *
* * *
```

---

## **🔹 Key Differences Between Both Approaches**
| Feature         | First Code (Numbers) | Second Code (Stars) |
|----------------|----------------------|----------------------|
| Output         | Prints `N` | Prints `*` |
| Number of Loops | Two (`O(N²)`) | One (`O(N)`) |
| Space Complexity | `O(1)` | `O(1)` |
| Efficiency | Slightly slower due to inner loop | Faster due to direct string multiplication |
| Printing Method | Uses `print(N, end=" ")` in a loop | Uses `print("* " * N)` |

---

## **🎯 Key Takeaways**
✅ **Use nested loops** if you need precise control over each printed element.  
✅ **Use string multiplication (`"* " * n`)** when printing uniform patterns efficiently.  
✅ **Both approaches work well**, but the second one is cleaner for simple patterns.
