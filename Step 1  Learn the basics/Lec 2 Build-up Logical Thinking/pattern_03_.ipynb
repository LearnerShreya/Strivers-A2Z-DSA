Pattern 12  
https://cdnpractice.geeksforgeeks.org/problems/double-triangle-pattern-1662664259/1

```python
class Solution:
    def printTriangle(self, N):
        # Code here
        for i in range(1, N+1):
            for j in range(1, i+1):
                print(j, end=" ")
            for j in range(N-i):
                print("  "*2, end="")
            for j in range(i, 0, -1):
                print(j, end=" ")
            print()

sol = Solution()
sol.printTriangle(5)
```

```python
class Solution:
    def printTriangle(self, N):
        for i in range(1, N+1):
            first_half = " ".join(str(j) for j in range(1, i+1))
            spaces = " " * (2 * (N - i))  # Adjusted spaces to maintain alignment
            second_half = " ".join(str(j) for j in range(i, 0, -1))
            print(first_half + spaces + second_half)

sol = Solution()
sol.printTriangle(5)
```

Pattern-19: Symmetric-Void Pattern  
https://takeuforward.org/pattern/pattern-19-symmetric-void-pattern/  
Pattern 19  
https://www.geeksforgeeks.org/problems/double-triangle-pattern/0

```python
class Solution:
    def printTriangle(self, N):
        # Upper half of the pattern
        for i in range(N):
            first_half = "*" * (N - i)
            space = " " * (2 * i)
            second_half = "*" * (N - i)
            print(first_half + space + second_half)

        # Lower half (mirror image of upper half)
        for i in range(N - 1, -1, -1):
            first_half = "*" * (N - i)
            space = " " * (2 * i)
            second_half = "*" * (N - i)
            print(first_half + space + second_half)

# Example usage
sol = Solution()
sol.printTriangle(5)
```

Pattern - 20: Symmetric-Butterfly Pattern  
https://takeuforward.org/pattern/pattern-20-symmetric-butterfly-pattern/  
Pattern 20  
https://www.geeksforgeeks.org/problems/double-triangle-pattern-1662287416/0

1. **Wrong loop range:**  
   - **Incorrect:** `for i in range(1, N):` → Starts from `1`, should start from `N-1` and go down.  
   - **Fix:** `for i in range(N - 1, 0, -1):`  

2. **Incorrect space calculation:**  
   - **Incorrect:** `space = " " * (2 * i)` → Increases incorrectly.  
   - **Fix:** `space = " " * (2 * (N - i))` → Ensures symmetry with the upper part.  

```python
class Solution:
    def printTriangle(self, N):
        # Code here
        for i in range(1, N+1):
            first_half = "*" * i
            space = " " * (2*(N-i))
            second_half = "*" * i
            print(first_half + space + second_half)
        for i in range(1, N):
            first_half = "*"*(N-i) 
            space = " " * (2*(i))
            second_half = "*"*(N-i)
            print(first_half + space + second_half)

sol = Solution()
sol.printTriangle(5)
```

Pattern - 21: Hollow Rectangle Pattern  
https://takeuforward.org/pattern/pattern-21-hollow-rectangle-pattern/  
Pattern 21  
https://www.geeksforgeeks.org/problems/square-pattern-1662287714/0

```python
class Solution:
    def printTriangle(self, N):
        # Code here
        for i in range(N):
            if i==0 or i==N-1:
                print("*"*N)
            else:
                print("*" + " "*(N-2) + "*")

sol = Solution()
sol.printTriangle(5)
```

Pattern - 22: The Number Pattern

Pattern 22

```python
class Solution:
    def printTriangle(self, N):
        # Code here
        for i in range(2*N-1):
            for j in range(2*N-1):
                val = N - min(i, j, size-i-1, size-j-1)
                print(val, end=" ")
            print()

sol = Solution()
sol.printTriangle(5)
```
