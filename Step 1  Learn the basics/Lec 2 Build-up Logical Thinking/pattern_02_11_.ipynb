https://www.geeksforgeeks.org/problems/right-triangle/1

```python
class Solution:
    def printTriangle(self, N):
        # Loop through rows from 1 to N
        for i in range(N):
            print("* " * (i + 1))  # Print (i+1) stars in each row

# Example usage:
sol = Solution()
sol.printTriangle(5)
```

Pattern 3

https://takeuforward.org/pattern/pattern-3-right-angled-number-pyramid/

https://www.geeksforgeeks.org/problems/triangle-number/1

```python
class Solution:
    def printTriangle(self, N):
        # for i in range(N+1):
        for i in range(1,N+1):   # Use (1, N+1) as it avoids an extra iteration (i=0), improves efficiency,
            for j in range(1,i+1):
                print(j, end=" ")
            print()

sol = Solution()
sol.printTriangle(5)
```

Pattern - 6: Inverted Numbered Right Pyramid  
https://takeuforward.org/pattern/pattern-6-inverted-numbered-right-pyramid/

```python
class Solution:
    def printTriangle(self, N):
        for i in range(N):  
            for j in range(1, N - i + 1):  # Start from 1 instead of 0
                print(j, end=" ")
            print()  # Move to the next line

sol = Solution()
sol.printTriangle(5)
```

```python
class Solution:
    def printTriangle(self, N):
        for i in range(1, N+1):
            for j in range(1, N-i+1):
                print(" ", end="")

            for j in range(1, 2*i):
                print("*", end="")
            print()

sol = Solution()
sol.printTriangle(5)
```

Pattern - 9: Diamond Star Pattern  
https://takeuforward.org/pattern/pattern-9-diamond-star-pattern/  
Pattern  
https://www.geeksforgeeks.org/problems/pattern/0

```python
class Solution:
    def printDiamond(self, N):
        # Code here
        for i in range(N,0,-1):
            print(" "*(i-1) + "* "*(N-i+1))
        for j in range(N,0,-1):
            print(" "*(N-j) + "* "*(j))

sol = Solution()
sol.printDiamond(5)
```

Pattern - 10: Half Diamond Star Pattern  
https://takeuforward.org/pattern/pattern-10-half-diamond-star-pattern/  
Pattern 10  
https://www.geeksforgeeks.org/problems/triangle-pattern-1661718013/0

```python
class Solution:
    def printTriangle(self, n):
        # Code here
        for i in range(1, n+1):
            print("* "*i, end="")
            print()
        for j in range(n-1,0,-1):
            print("* "*j, end="")
            print()

sol = Solution()
sol.printTriangle(5)
```

Pattern - 11: Binary Number Triangle Pattern  
https://takeuforward.org/pattern/pattern-11-binary-number-triangle-pattern/

```python
class Solution:
    def printTriangle(self, N):
        # Code here
        for i in range(1,N+1):
            for j in range(i):
                print((i+j)%2, end=" ")
            print()
sol = Solution()
sol.printTriangle(5)
```
